```python
import numpy as np
import matplotlib.pyplot as plt
```

## 油田の数

```python
def gen():
    n = np.random.randint(10, 20+1)
    m = np.random.randint(2, (n*n//20)+1)
    return m

n_samples = 1_000_000
samples = [gen() for _ in range(n_samples)]
```

```python
hist = plt.hist(samples, bins=range(2,21+1))
```

```python
occ = hist[0]
s = 0
sum = []
for i in range(len(occ)):
    s += occ[i]
    sum.append(s)
sum = np.array(sum)
prob = sum/n_samples
```

```python
plt.plot(list(range(2,21)), prob, "o")
plt.ylim(0, 1)
```

## 観測誤差

```python
eps_base = 0.01
eps = eps_base * np.arange(1,21)
sigma = np.sqrt(eps * (1 - eps))
```

```python
sigma
```

```python
plt.plot(eps, sigma, "o")
plt.xlim(0,0.2)
plt.ylim(0,0.5)
```

```python
area_range = np.arange(1, 20+1)
eps_grid, area_grid = np.meshgrid(eps, area_range)
```

```python
sigma_map = np.sqrt(area_grid * eps_grid * (1 - eps_grid))
```

```python
plt.contour(eps_grid, area_grid, sigma_map, levels=20)
```
